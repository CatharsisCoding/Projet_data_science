```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd

# Lire le fichier StepOne_LabelTrain.csv
df_stepone = pd.read_csv('StepOne_LabelTrain.csv')

# Trouver une ligne o√π seulement "Washing Machine" a une valeur de 1 et tous les autres appareils ont une valeur de 0
washing_machine_only = df_stepone.index[(df_stepone["Washing Machine"] == 1) & 
                                       (df_stepone["Dishwasher"] == 0) & 
                                       (df_stepone["Tumble Dryer"] == 0) & 
                                       (df_stepone["Microwave"] == 0) & 
                                       (df_stepone["Kettle"] == 0)].values[0]

dishwasher_only = df_stepone.index[(df_stepone["Washing Machine"] == 0) & 
                                       (df_stepone["Dishwasher"] == 1) & 
                                       (df_stepone["Tumble Dryer"] == 0) & 
                                       (df_stepone["Microwave"] == 0) & 
                                       (df_stepone["Kettle"] == 0)].values[0]
tumble_only = df_stepone.index[(df_stepone["Washing Machine"] == 0) & 
                                       (df_stepone["Dishwasher"] == 0) & 
                                       (df_stepone["Tumble Dryer"] == 1) & 
                                       (df_stepone["Microwave"] == 0) & 
                                       (df_stepone["Kettle"] == 0)].values[0]
microwave_only = df_stepone.index[(df_stepone["Washing Machine"] == 0) & 
                                       (df_stepone["Dishwasher"] == 0) & 
                                       (df_stepone["Tumble Dryer"] == 0) & 
                                       (df_stepone["Microwave"] == 1) & 
                                       (df_stepone["Kettle"] == 0)].values[0]
kettle_only = df_stepone.index[(df_stepone["Washing Machine"] == 0) & 
                                       (df_stepone["Dishwasher"] == 0) & 
                                       (df_stepone["Tumble Dryer"] == 0) & 
                                       (df_stepone["Microwave"] == 0) & 
                                       (df_stepone["Kettle"] == 1)].values[0]


```

```python
tab_final=[washing_machine_only,dishwasher_only,tumble_only,microwave_only,kettle_only]
tab_text=["washing_machine_only","dishwasher_only","tumble_only","microwave_only","kettle_only"]
# Lire le fichier CSV
df_train = pd.read_csv('InputTrain.csv')

# Afficher l'√©volution de ["washing_machine_only","dishwasher_only","tumble_only","microwave_only","kettle_only"]

for elem,a in zip(tab_final,tab_text):
    plt.plot(df_train.iloc[elem, 2:])
    plt.title('Evolution de '+a)
    plt.xlabel('Index des pas de temps')
    plt.ylabel('Consommation en watts')


    x_ticks = np.arange(0, len(df_train.columns[2:]), 300)

    # Sp√©cifier les valeurs des x sur l'axe des abscisses
    plt.xticks(x_ticks,rotation=90)
    plt.show()


```
